```python
import matplotlib.pyplot as plt

# Data for the plot
dimensions = [112, 196, 256, 392, 512]
full = [0.73, 0.70, 0.7111, 0.76, 0.78]
samm = [0.76, 0.78, 0.8124, 0.80, 0.82]
casme_ii = [0.88, 0.90, 0.9516, 0.92, 0.94]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the three lines
plt.plot(dimensions, full, label='Full', marker='o', linestyle='-', color='blue')
plt.plot(dimensions, samm, label='SAMM', marker='o', linestyle='-', color='green')
plt.plot(dimensions, casme_ii, label='CASME II', marker='o', linestyle='-', color='red')

# Labeling the axes and the plot
plt.xlabel('Dimensions', fontsize=12)
plt.ylabel('UAR', fontsize=12)
plt.title('UAR vs Dimensions for Full, SAMM, and CASME II', fontsize=14)

# Add a legend to differentiate between the three lines
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
