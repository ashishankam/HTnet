```python
import matplotlib.pyplot as plt

# Data
dimensions = [112, 196, 256, 392, 512]
uf1_full = [0.64, 0.72, 0.7358, 0.68, 0.65]
uf1_samm = [0.78, 0.80, 0.8131, 0.72, 0.71]
uf1_casme2 = [0.90, 0.92, 0.9532, 0.84, 0.86]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(dimensions, uf1_full, marker='o', label='Full')
plt.plot(dimensions, uf1_samm, marker='o', label='SAMM')
plt.plot(dimensions, uf1_casme2, marker='o', label='CASME II')

# Adding labels, title, and legend
plt.xlabel('Dimensions', fontsize=12)
plt.ylabel('UF1', fontsize=12)
plt.title('UF1 vs Dimensions for Full, SAMM, and CASME II', fontsize=14)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
